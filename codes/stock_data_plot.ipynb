```python
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
start_date = '2013-01-01'
end_date = '2016-07-01'
ticker = 'AAPL TSLA META AMZN MSFT GOOG GE DB'
```

```python
data = yf.download(ticker, start_date, end_date)
print(data.tail())
```

```python
data.info()
```

```python
# Show columns name
data.columns.tolist()
```

```python
data[('Open', 'AAPL')].index
```

```python
plt.figure(figsize=(8,4), dpi=300)
plt.title("Stock Data (High) of Different Companies ")


for i in list(ticker.split(" ")):
    plt.plot(data[('High', i)].index, data[('High', i)].values, label=str(i))
    plt.legend()

plt.savefig('stockprice.jpg')
```
